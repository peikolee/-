```python
# User handles
df['Clean_Text'] = df['Text'].apply(nfx.remove_userhandles)
```

```python
# Stopwords
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_stopwords)
```

```python
import re
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def preprocess_text(text):
    # 移除用户名handle
    text = nfx.remove_userhandles(text)
    
    # 移除停用词
    text = nfx.remove_stopwords(text)
    
    # 小写化
    text = text.lower()
    
    # 移除标点符号
    text = re.sub(r'[^\w\s]', '', text)
    
    # 词干提取
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    
    # 其他预处理步骤
    # ...
    
    return text
```

```python
import random
from nltk.corpus import wordnet

def augment_data(X, y):
    X_augmented = []
    y_augmented = []
    
    for i, text in enumerate(X):
        tokens = text.split()
        augmented_tokens = []
        for token in tokens:
            if random.random() < 0.1:  # 10% 的替换概率
                synsets = wordnet.synsets(token)
                if synsets:
                    synonym = random.choice(synsets).lemmas()[0].name()
                    augmented_tokens.append(synonym)
                else:
                    augmented_tokens.append(token)
            else:
                augmented_tokens.append(token)
        augmented_text = ' '.join(augmented_tokens)
        X_augmented.append(augmented_text)
        y_augmented.append(y[i])
    
    return X_augmented, y_augmented
```

```python
# Features & Labels
Xfeatures = df['Clean_Text']
ylabels = df['Emotion']
```

# It is advisable to split before applying pipelines because it prevents data leakage.

```python
#  Split Data
x_train,x_test,y_train,y_test = train_test_split(Xfeatures,ylabels,test_size=0.3,random_state=42)
```

```python
# Build Pipeline
from sklearn.pipeline import Pipeline
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer

```

```python
# LogisticRegression Pipeline
from sklearn.linear_model import LogisticRegression

pipe_lr = Pipeline(steps=[('cv', CountVectorizer()),
                         ('lr', LogisticRegression(C=1.0, penalty='l2'))])
# 训练随机森林模型 
from sklearn.ensemble import RandomForestClassifier
pipe_rf = Pipeline(steps=[('cv', CountVectorizer()),
                         ('rf', RandomForestClassifier(n_estimators=100, max_depth=5))])
```

```python

def cross_validate(X, y, model, n_splits=5):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    scores = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Apply data augmentation
        X_train_augmented, y_train_augmented = augment_data(X_train, y_train)

        # Train the model
        model.fit(X_train_augmented, y_train_augmented)

        # Evaluate the model on the test set
        y_pred = model.predict(X_test)
        score = accuracy_score(y_test, y_pred)
        scores.append(score)

    # Calculate the average accuracy
    mean_score = np.mean(scores)
    print(f"Average Accuracy: {mean_score:.4f}")

    return mean_score
```

```python
# Train and Fit Data
pipe_lr.fit(x_train, y_train)
pipe_rf.fit(x_train, y_train)
```

```python
# Check Accuracy
lr_score = pipe_lr.score(x_test, y_test)
print(f"Logistic Regression Accuracy: {lr_score:.2f}")

rf_score = pipe_rf.score(x_test, y_test)
print(f"Random Forest Accuracy: {rf_score:.2f}")
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt

# Logistic Regression Pipeline
train_sizes, train_scores, valid_scores = learning_curve(
    pipe_lr, Xfeatures, ylabels, cv=5,
    scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10))

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Training Score')
plt.plot(train_sizes, np.mean(valid_scores, axis=1), label='Validation Score')
plt.xlabel('Training Set Size')
plt.ylabel('Score')
plt.title('Learning Curve - Logistic Regression')
plt.legend()
plt.show()

# Random Forest Pipeline
train_sizes, train_scores, valid_scores = learning_curve(
    pipe_rf, Xfeatures, ylabels, cv=5,
    scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10))

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Training Score')
plt.plot(train_sizes, np.mean(valid_scores, axis=1), label='Validation Score')
plt.xlabel('Training Set Size')
plt.ylabel('Score')
plt.title('Learning Curve - Random Forest')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt

# Logistic Regression Validation Curve
param_range = np.logspace(-3, 3, 5)
train_scores, valid_scores = validation_curve(
    pipe_lr, Xfeatures, ylabels, param_name='lr__C', param_range=param_range,
    cv=5, scoring='accuracy')

plt.figure(figsize=(8, 6))
plt.plot(param_range, np.mean(train_scores, axis=1), label='Training Score')
plt.plot(param_range, np.mean(valid_scores, axis=1), label='Validation Score')
plt.xlabel('Regularization Parameter (C)')
plt.ylabel('Score')
plt.title('Validation Curve - Logistic Regression')
plt.xscale('log')
plt.legend()
plt.show()

# Random Forest Validation Curve
param_range = np.linspace(10, 200, 5, dtype=int)
train_scores, valid_scores = validation_curve(
    pipe_rf, Xfeatures, ylabels, param_name='rf__n_estimators', param_range=param_range,
    cv=5, scoring='accuracy')

plt.figure(figsize=(8, 6))
plt.plot(param_range, np.mean(train_scores, axis=1), label='Training Score')
plt.plot(param_range, np.mean(valid_scores, axis=1), label='Validation Score')
plt.xlabel('Number of Estimators')
plt.ylabel('Score')
plt.title('Validation Curve - Random Forest')
plt.legend()
plt.show()
```
